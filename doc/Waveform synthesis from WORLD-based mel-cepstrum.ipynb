```julia
using VoiceConversion
using PyPlot
```

```julia
using WAV
data, fs = wavread("test.wav")
data = vec(data) # monoral

# Visualize the speech signal in time-domain
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(data), data, label="a speech signal")
xlim(1, endof(data))
xlabel("sample")
legend()
```

```julia
using WORLD

w = World(fs=fs, period=5.0)
```

```julia
# Fundamental frequency (f0) estimation by DIO
f0, timeaxis = dio1(w, data)  

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(timeaxis, f0, label="F0 trajectory estimated by DIO")

# F0 refinement by StoneMask
f0 = stonemask(w, data, timeaxis, f0)

plot(timeaxis, f0, label="Refined f0 by StoneMask")
xlabel("Time [sec]")
ylabel("Frequency  [Hz]")
legend()
```

```julia
# Spectral envelope estimation by CheapTrick
spectrogram = cheaptrick(w, data, timeaxis, f0)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(spectrogram), origin="lower", aspect="auto")
```

```julia
# Aperiodicity ration 
estimation
aperiodicity = aperiodicityratio(w, data, f0, timeaxis)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(aperiodicity, origin="lower", aspect="auto")
```

```julia
# Spectrum envelope -> Mel-Ceptrum
using MCepAlpha
order = 50
alpha = mcepalpha(fs)
melspectrogram = wsp2mc(spectrogram, order, alpha)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(melspectrogram, origin="lower", aspect="auto")
colorbar()
```

```julia
# Mel-cepstrum -> Spectral envelope
reconstructed_spectrogram = mc2wsp(melspectrogram, size(spectrogram, 1), -alpha)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(reconstructed_spectrogram), origin="lower", aspect="auto")
```

```julia
norm(spectrogram - reconstructed_spectrogram)/norm(spectrogram)
```

```julia
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(spectrogram - reconstructed_spectrogram, origin="lower", aspect="auto")
colorbar()
```

```julia
# Sysnthesis from f0, spectral envelope and aperiodicity ratio.
synthesizedA = synthesis_from_aperiodicity(w, f0, spectrogram, aperiodicity, length(data))

println("Mean error: ", mean(abs(synthesizedA-data))/maxabs(data))

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(synthesizedA), synthesizedA, "r-+", label="reconstructed speech signal from original spectral envelope     ")
plot(1:endof(data), data, label="original speech signal")
xlim(1, endof(synthesizedA))
xlabel("sample")
legend()
```

```julia
# Sysnthesis from f0, spectral envelope and aperiodicity ratio.
synthesizedAA = synthesis_from_aperiodicity(w, f0, reconstructed_spectrogram, aperiodicity, length(data))

println("Mean error: ", mean(abs(synthesizedAA-data))/maxabs(data))

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(synthesizedAA), synthesizedAA, "r-+", label="reconstructed speech signal from mel-cepstrum")
plot(1:endof(data), data, label="original speech signal")
xlim(1, endof(synthesizedAA))
xlabel("sample")
legend()
```
