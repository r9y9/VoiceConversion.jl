## WORLD-based Mel-Cesptrum

This notebook shows how to compute mel-cesptrum from WORLD-based spectral envelope with some helpful visualizations.

```julia
using VoiceConversion
using WORLD
using WAV
using PyPlot
using SPTK
```

```julia
data, fs = wavread(joinpath(Pkg.dir("VoiceConversion"), "test", "data", "test16k.wav"))
data = vec(data) # monoral

# Visualize the speech signal in time-domain
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(data), data, label="a speech signal")
xlim(1, endof(data))
xlabel("sample")
legend()
```

```julia
period = 5.0
order = 25
alpha = 0.35
println("$(maximum(data)) and $(minimum(data))")
```

```julia
w = World(fs, period)
```

```julia
# Fundamental frequency (f0) estimation by DIO
f0, timeaxis = dio(w, data)
    
# F0 re-estimation by StoneMask
f0 = stonemask(w, data, timeaxis, f0)
    
# Spectral envelope estimation
spectrogram = cheaptrick(w, data, timeaxis, f0)
println(size(spectrogram))
```

```julia
# Spectrogram visualization
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(spectrogram), origin="lower", aspect="auto", interpolation="nearest")
colorbar()
```

```julia
# Pick a spectrum
spec = spectrogram[:,50]

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(spec)
xlim(0, 512)
```

```julia
# Symmetrized spectrum for convinience
symmetrized = [spec, reverse(spec[2:end])]

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(symmetrized)
xlim(0, 1024)
```

```julia
# Log
logspec = log(symmetrized)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(logspec)
xlim(0, 1024)
```

```julia
# Log spectrum -> Mel-cepstrum
mc = logamp2mc(logspec, order, alpha)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(mc)
xlim(0, order)
```

```julia
# WORLD spectrum to mel-cepstrum onversion
mcgram = wsp2mc(spectrogram, order, alpha)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(mcgram, origin="lower", aspect="auto", interpolation="nearest")
colorbar()
```
