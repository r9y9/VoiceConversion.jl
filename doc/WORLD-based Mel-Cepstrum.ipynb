## WORLD-based Mel-Cesptrum

This notebook shows how to compute mel-cesptrum from WORLD-based spectral envelope with some helpful visualizations.

```julia
using WORLD
using WAV
```

```julia
using PyPlot
```

```julia
# Read test data
x, fs = wavread("test16k.wav", format="int")
@assert size(x, 2) == 1 "The input data must be monoral." 
x = float64(x[:])
fs = int(fs)
```

```julia
period = 5.0
order = 25
alpha = 0.35
println("$(maximum(x)) and $(minimum(x))")
```

```julia
w = World(fs=fs, period=period)
```

```julia
# Fundamental frequency (f0) estimation by DIO
f0, timeaxis = dio1(w, x)
    
# F0 re-estimation by StoneMask
f0 = stonemask(w, x, timeaxis, f0)
    
# Spectral envelope estimation
spectrogram = cheaptrick(w, x,timeaxis, f0)
println(size(spectrogram))
```

```julia
# Spectrogram visualization
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(spectrogram), origin="lower", aspect="auto", interpolation="nearest")
colorbar()
```

```julia
# Pick a spectrum
spec = spectrogram[:,50]

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(spec)
xlim(0, 512)
```

```julia
# Symmetrized spectrum for convinience
symmetrized = [spec, reverse(spec[2:end])]

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(symmetrized)
xlim(0, 1024)
```

```julia
import SPTK
const sptk = SPTK
```

```julia
# Log
logspec = log(symmetrized)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(logspec)
xlim(0, 1024)
```

```julia
# logamp2mcep converts log-amplitude spectrum to mel-cepstrum.
function logamp2mcep(logamp::Vector{Float64}, order::Int, alpha::Float64)
    ceps = real(ifft(logamp))
    ceps[1] /= 2.0
    return sptk.freqt(ceps, order, alpha)
end
```

```julia
# Log spectrum -> Mel-cepstrum
mc = logamp2mcep(logspec, order, alpha)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(mc)
xlim(0, order)
```

```julia
mcgram = zeros(order+1, size(spectrogram, 1))

for i=1:size(spectrogram, 1)
    spec = spectrogram[:,i]
    symmetrized = [spec, reverse(spec[2:end])]
    logspec = log(symmetrized)
    mcgram[:,i] = logamp2mcep(logspec, order, alpha)
end
```

```julia
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(mcgram, origin="lower", aspect="auto", interpolation="nearest")
colorbar()
```
